## 1. Introducción a Streamlit

Streamlit es una biblioteca de Python que simplifica el desarrollo de aplicaciones de datos. Las aplicaciones se escriben en un archivo `.py` con código de Python puro, y al ejecutar `streamlit run nombre_del_script.py` se lanza automáticamente un servidor web que muestra la app.

Características clave:

- **Fácil de instalar y ejecutar**. Con `pip install streamlit` y una sola línea de comando (`streamlit run app.py`) puedes poner en marcha tu aplicación【814966532335†L90-L102】.
- **Interactividad sencilla**. Permite "hornear" interactividad en las apps mediante **botones, sliders, entradas de texto y más**【338445402297917†L235-L238】. Declarar un widget es tan sencillo como asignar el resultado de una función (`st.slider`, `st.selectbox`, etc.) a una variable.
- **Sin conocimientos de front‑end**. No necesitas programar el backend ni diseñar HTML; Streamlit se encarga de todo【475725854611736†L125-L131】.
- **Uso de Python nativo**. Puedes utilizar tus bibliotecas favoritas de análisis (`pandas`, `numpy`, `matplotlib`, `plotly`) y mostrarlas directamente con funciones como `st.dataframe` o `st.chart`.



## 2. Widgets e interactividad

Streamlit ofrece una amplia variedad de **widgets de entrada** que permiten a los usuarios interactuar con la aplicación. Estos widgets permiten filtrar, seleccionar y parametrizar la visualización de datos. Algunos de los más usados:

- **st.button**: crea un botón que devuelve `True` cuando se presiona.
- **st.slider**: deslizador para elegir un valor numérico o un rango.
- **st.text_input** y **st.number_input**: campos de texto y número.
- **st.selectbox** y **st.radio**: selección de una opción de una lista.

La documentación de Streamlit resalta que estos widgets te permiten integrar interactividad en tus apps de forma directa【338445402297917†L235-L238】. A continuación se presenta un ejemplo de cómo usar un *slider* y un *selectbox* para filtrar datos.


````python
# Ejemplo: widgets de Streamlit
import streamlit as st
import pandas as pd
import seaborn as sns

# Cargar un conjunto de datos de ejemplo
df = sns.load_dataset('penguins')

st.title('Ejemplo de filtros con Streamlit')

# Crear un selectbox para escoger la especie
especie = st.selectbox('Selecciona la especie', df['species'].unique())

# Crear un slider para elegir un rango de masa corporal
masa_min, masa_max = st.slider('Rango de masa corporal (g)',
                               min_value=float(df['body_mass_g'].min()),
                               max_value=float(df['body_mass_g'].max()),
                               value=(float(df['body_mass_g'].min()), float(df['body_mass_g'].max())))

# Filtrar datos según los valores seleccionados
df_filtrado = df[(df['species'] == especie) & (df['body_mass_g'] >= masa_min) & (df['body_mass_g'] <= masa_max)]

# Mostrar una tabla con los datos filtrados
st.dataframe(df_filtrado)
````


En este ejemplo:

- `st.selectbox` crea un menú desplegable con las especies disponibles.
- `st.slider` permite seleccionar un rango de valores para la masa corporal. 
- Los datos se filtran en tiempo real, y el resultado se muestra en un `st.dataframe`.

> **Nota:** Para ejecutar una app de Streamlit desde un cuaderno de Jupyter se recomienda guardar el código en un archivo `.py` y correrlo con `streamlit run archivo.py`. Aquí mostramos el código para ilustrar su estructura.



## 3. Layout y contenedores

Controlar la organización de los elementos es clave para crear dashboards bien estructurados. Streamlit proporciona varias formas de distribuir los componentes en la interfaz【820724192007135†L175-L178】:

- **st.sidebar**: una barra lateral donde puedes colocar widgets y menús de navegación para mantener limpia la vista principal. Según la API, se pueden usar dos estilos: notación por atributo (`st.sidebar.selectbox(...)`) o utilizando un contexto `with`【355072310659934†L182-L204】.
- **st.columns**: divide la página en columnas de ancho variable, permitiendo colocar gráficos o tablas lado a lado.
- **st.expander**: un contenedor plegable para ocultar detalles que no son esenciales a primera vista.
- **st.tabs**: crea pestañas para organizar diferentes secciones.

A continuación se muestra un ejemplo de layout usando la barra lateral y las columnas.


````python

# Ejemplo: uso de st.sidebar y columnas
import streamlit as st
import pandas as pd
import numpy as np

st.title('Dashboard con barra lateral y columnas')

# Widgets en la barra lateral
with st.sidebar:
    st.header('Opciones de filtrado')
    rango = st.slider('Selecciona un rango', 0, 100, (25, 75))
    opcion = st.radio('Mostrar gráficos', ['Histograma', 'Dispersión'])

# Generar datos aleatorios
np.random.seed(0)
datos = np.random.randn(500)
datos_filtrados = datos[(datos >= (rango[0] - 50) / 10) & (datos <= (rango[1] - 50) / 10)]

# Layout en columnas
a, b = st.columns(2)

# Mostrar resultados en columnas
with a:
    st.subheader('Datos filtrados')
    st.write(pd.DataFrame({'Valores': datos_filtrados}))

with b:
    if opcion == 'Histograma':
        st.subheader('Histograma')
        st.hist_chart(datos_filtrados)
    else:
        st.subheader('Diagrama de dispersión')
        st.scatter_chart(pd.DataFrame({'x': np.arange(len(datos_filtrados)), 'y': datos_filtrados}))
````


En este ejemplo:

- La barra lateral (definida dentro del contexto `with st.sidebar`) contiene un slider y un radio botón para configurar los parámetros del dashboard【355072310659934†L217-L232】.
- En la vista principal, `st.columns` divide la pantalla en dos columnas. La primera muestra una tabla con los datos filtrados; la segunda muestra un histograma o un diagrama de dispersión según la opción seleccionada.

Puedes experimentar cambiando los valores y observar cómo se actualizan los gráficos en tiempo real.



## 4. Caché y rendimiento

Cuando tu aplicación necesita cargar datos pesados o ejecutar funciones costosas, es recomendable usar el decorador `@st.cache_data` (o `@st.cache_resource`) para almacenar en memoria los resultados y evitar recomputaciones innecesarias. Por ejemplo:


````python

import streamlit as st
import pandas as pd
import time

@st.cache_data
def cargar_datos(path):
    # Simula una operación costosa
    time.sleep(3)
    return pd.read_csv(path)

st.write('Cargando datos...')
df = cargar_datos('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
st.success('Datos cargados')
st.write(df.head())
````


Al usar `@st.cache_data`, la primera vez que se llama a `cargar_datos` se realiza la operación (por ejemplo, leer un archivo o consultar una base de datos). Las siguientes veces, Streamlit recupera el resultado almacenado en caché, reduciendo el tiempo de respuesta.
