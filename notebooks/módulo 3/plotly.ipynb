# Módulo 3: Salto a la interactividad con Plotly

En este módulo aprenderás a crear **gráficos interactivos** con la biblioteca Plotly. 
Mientras que Matplotlib y Seaborn generan gráficos estáticos, Plotly permite 
interactuar con los datos mediante *hover*, *zoom*, filtros y animaciones. 
Los notebooks de Jupyter son un entorno ideal para combinar código, datos, explicación y gráficos.

Plotly proporciona una API en Python que se integra perfectamente con Jupyter. 
Al llamar a `iplot` o utilizar las funciones de Plotly Express dentro de un notebook, 
los gráficos se muestran como un iframe interactivo en la misma página. Según la documentación oficial, 
`iplot` genera automáticamente una versión interactiva del gráfico dentro del notebook, 
lo que te permite mantener el análisis y las visualizaciones en un único lugar∫.

Importamos las librerías necesarias y cargamos algunos conjuntos de datos de ejemplo incluidos en Plotly.

````python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
````

## Plotly Express: gráficos básicos e interacción

[Plotly Express](https://plotly.com/python/plotly-express/) es la interfaz de alto nivel de Plotly. 
Permite generar gráficos interactivos con pocas líneas de código. 
Las funciones de Plotly Express aceptan DataFrames de Pandas y te permiten controlar 
el mapeo de variables a ejes, colores, tamaños, facetas, etc.

A continuación construiremos un diagrama de dispersión interactivo con el conjunto 
de datos `tips`. Utilizaremos diferentes variables para colorear y dimensionar los puntos, 
además de crear facetas por día de la semana. Puedes pasar el ratón por encima de cada punto 
para ver la información exacta, hacer *zoom* con el ratón o seleccionar rangos con la herramienta de 
selección.

````python
# Cargar el conjunto de datos 'tips'
df = px.data.tips()
df
````

````python
# Crear un diagrama de dispersión interactivo
fig = px.scatter(df, x='total_bill', y='tip', color='smoker', size='size',
                 facet_col='day',
                 title='Relación entre la cuenta total y la propina')
# Mostrar el gráfico
fig.show()
````

En el diagrama anterior cada punto representa una comida (servicio por mesa) en un restaurante. 
Las facetas muestran los distintos días de la semana, el color indica si el cliente fumaba 
y el tamaño del punto refleja el número de comensales. Al pasar el cursor por encima, 
aparecen las variables de esa observación. Este tipo de interacción es una de las 
principales ventajas de Plotly frente a bibliotecas estáticas.

## Animaciones y series temporales

Plotly facilita la creación de animaciones, en las que cada fotograma representa una 
porción diferente del conjunto de datos. Usaremos el conjunto `gapminder` para 
visualizar la evolución del PIB y la esperanza de vida por país a lo largo del tiempo.

````python
# Cargar datos de Gapminder
gapminder = px.data.gapminder()
````

````python
gapminder
````

````python
# Definir rangos fijos con un pequeño margen para evitar recortes en la animación
x_min, x_max = gapminder['gdpPercap'].min(), gapminder['gdpPercap'].max()
y_min, y_max = gapminder['lifeExp'].min(), gapminder['lifeExp'].max()
range_x = [x_min * 0.9, x_max * 1.1]
range_y = [y_min - 2, y_max + 2]

# Crear un scatter animado
fig_anim = px.scatter(gapminder, x='gdpPercap', y='lifeExp',
                       animation_frame='year', animation_group='country',
                       size='pop', color='continent', hover_name='country',
                       log_x=True, range_x=range_x, range_y=range_y,
                       size_max=60,
                       title='Evolución del PIB per cápita y esperanza de vida (Gapminder)')
fig_anim.show()
````

Puedes mover la barra de tiempo o hacer clic en el botón de reproducción para 
ver cómo cambian las posiciones de los países a través de los años. Además, es posible hacer 
*zoom* y arrastrar en cualquier momento de la animación.

## Gráficos financieros: velas japonesas

Plotly también incluye componentes para visualizar series financieras. Las velas japonesas 
(candlesticks) muestran precios de apertura, máximo, mínimo y cierre en un periodo. 
Utilizaremos el conjunto `stocks` para visualizar los datos de una acción.

````python
# Datos de ejemplo de acciones
stocks = px.data.stocks()
stocks
````

````python
# Seleccionar las primeras fechas de una acción (por ejemplo Apple)
df_stock = stocks[['date', 'AAPL']].copy()
df_stock['date'] = pd.to_datetime(df_stock['date'])

# Generar columnas OHLC sintéticas a partir de los precios de cierre
df_stock['open'] = df_stock['AAPL'].shift(1)
df_stock['high'] = df_stock[['AAPL','open']].max(axis=1) + 2
df_stock['low'] = df_stock[['AAPL','open']].min(axis=1) - 2
df_stock['close'] = df_stock['AAPL']
df_stock = df_stock.dropna()

fig_candle = go.Figure(go.Candlestick(x=df_stock['date'],
                                      open=df_stock['open'],
                                      high=df_stock['high'],
                                      low=df_stock['low'],
                                      close=df_stock['close']))
fig_candle.update_layout(title='Ejemplo de gráfico de velas japonesas',
                      xaxis_title='Fecha',
                      yaxis_title='Precio (USD)',
                      xaxis_rangeslider_visible=False)
fig_candle.show()
````

El gráfico de velas permite identificar rápidamente las tendencias de precios en un periodo. 
Puedes seleccionar rangos de fechas con el ratón para ampliar el zoom o desplazarte por la serie.

## Diagramas de flujo: Sankey

Los diagramas de Sankey permiten visualizar el flujo de cantidades entre distintas categorías. 
A continuación construimos un ejemplo sencillo en el que se muestran flujos de energía.

````python
# Crear un diagrama de Sankey
sankey_data = dict(
    type='sankey',
    node=dict(label=['Generación','Residencial','Industrial','Transporte',
                     'Pérdidas'],
              pad=15, thickness=20, line=dict(color='black', width=0.5)),
    link=dict(source=[0,0,0,0], target=[1,2,3,4], value=[30, 25, 20, 10])
)

fig_sankey = go.Figure(data=[sankey_data])
fig_sankey.update_layout(title_text='Flujo de energía por sector', font_size=12)
fig_sankey.show()
````

En el ejemplo anterior, cada barra representa un nodo y los enlaces indican la proporción de 
energía que fluye hacia cada sector. Puedes pasar el cursor sobre las barras y enlaces para ver 
los valores exactos.

## Visualización geoespacial

Plotly facilita la creación de mapas interactivos sin necesidad de claves API. 
Usaremos `scatter_geo` para mostrar la esperanza de vida por país en el año 2007.

````python
# Filtrar el conjunto de datos para el año 2007
gapminder_2007 = gapminder[gapminder['year'] == 2007]

# Crear un mapa geoespacial interactivo
fig_geo = px.scatter_geo(gapminder_2007, locations='iso_alpha',
                         color='continent', size='pop',
                         hover_name='country',
                         projection='natural earth',
                         title='Esperanza de vida y población por país (2007)')
fig_geo.show()
````

Al mover el cursor sobre cada punto verás el nombre del país y sus valores de 
esperanza de vida y población. También puedes acercar o alejar la vista con la rueda del ratón.

## Dashboards interactivos y alternativas

Plotly permite crear gráficos interactivos directamente en un notebook Jupyter, lo que es ideal para 
análisis exploratorios. Tal y como explica la guía oficial, `iplot` genera una versión interactiva 
del gráfico dentro del notebook【34001796725094†L227-L235】, de manera que puedes mantener código, datos y visualización en un 
sólo documento.

Sin embargo, cuando se necesita construir **aplicaciones interactivas complejas** con múltiples 
controles (por ejemplo, desplegables, deslizadores, filtrado cruzado entre gráficos), los notebooks se quedan 
cortos. En esos casos conviene utilizar frameworks específicos de aplicaciones web como **Dash**, **Voila**, 
**Panel** o **Streamlit**.

* **Dash:** es un framework de Plotly para crear aplicaciones web analíticas completas. Está pensado 
para usarse fuera de Jupyter; expone más detalles de HTML/CSS/JS y puede requerir conocimientos 
de desarrollo web. Según un análisis de Quansight, Dash está diseñado para construir dashboards 
independientes y muchos usuarios lo encuentran difícil de mantener sin experiencia frontend【488324625742431†L229-L233】. 
Las primeras versiones no funcionaban en Jupyter, pero ahora existe una extensión llamada 
**JupyterDash** que permite ejecutar aplicaciones Dash en notebooks【488324625742431†L213-L241】. 

* **JupyterDash:** hace que sea sencillo desarrollar apps Dash en Jupyter. La guía de lanzamiento 
explica que basta con reemplazar `dash.Dash` por `jupyter_dash.JupyterDash` y llamar a 
`run_server` con distintos modos. El modo `inline` muestra la aplicación en un iframe dentro del notebook, 
permitiendo iterar rápidamente【421596776760889†L149-L169】. Además, JupyterDash ejecuta el servidor en un hilo 
de fondo para que el notebook no se bloquee【421596776760889†L134-L142】.

* **Voila:** fue diseñado para convertir notebooks en dashboards. El artículo de Quansight indica que 
cualquier notebook puede convertirse en dashboard con Voila, pero es necesario suprimir la salida que no 
quieras mostrar【488324625742431†L251-L255】. Voila es ideal para compartir análisis sin exponer el código. 
Eso sí, los dashboards complejos con múltiples interacciones son más difíciles de construir.

* **Panel:** ofrece una API imperativa y otra declarativa para crear aplicaciones con múltiples páginas y 
flujos. Puede usarse dentro de Jupyter o desde scripts. Quansight señala que Panel es flexible y se 
integra con la mayoría de bibliotecas de visualización y big data【488324625742431†L266-L279】.

* **Streamlit:** transforma scripts de Python en aplicaciones interactivas con un flujo lineal. Se 
desarrolla fuera del notebook; para publicar un notebook en Streamlit hay que convertirlo a script y 
eliminar las magias de Jupyter【488324625742431†L309-L314】.

En resumen, **los notebooks de Jupyter son adecuados para gráficos interactivos individuales y animaciones**, 
y se pueden ampliar con `JupyterDash` para prototipar dashboards sencillos. Si tu objetivo es construir 
aplicaciones ricas con múltiples controles y soportar muchos usuarios, considera migrar tu código a 
Dash, Panel, Voila o Streamlit, dependiendo de tus preferencias y del equipo con el que trabajes【488324625742431†L329-L335】.

### Ejemplo opcional: aplicación Dash en un notebook con JupyterDash

**Nota importante:** A partir de 2024, `JupyterDash` ha sido deprecado. La biblioteca Dash ahora 
soporta nativamente notebooks de Jupyter sin necesidad de JupyterDash. Para aplicaciones Dashboard 
complejas, se recomienda desarrollarlas como aplicaciones standalone usando `Dash` regular.


Para interactividad simple dentro de notebooks, puedes usar widgets de IPython/Jupyter (`ipywidgets`) anteriores (hover, zoom, animaciones, etc.).
o simplemente aprovechar las capacidades interactivas nativas de Plotly como se muestra en los ejemplos 

````python
# Ejemplo simple de interactividad con ipywidgets
# Este es el enfoque recomendado para interactividad simple en notebooks

try:
    from ipywidgets import interact, Dropdown
    import plotly.graph_objects as go
    
    df_tips = px.data.tips()
    
    def plot_scatter(color_col):
        fig = px.scatter(df_tips, x='total_bill', y='tip', color=color_col,
                        title=f'Dispersión propinas vs. cuenta total (color: {color_col})')
        fig.show()
    
    # Crear widget interactivo
    interact(plot_scatter, 
             color_col=Dropdown(options=['smoker', 'sex', 'time', 'day'], 
                               value='smoker', 
                               description='Color:'))
    
except ImportError:
    print("Para usar este ejemplo, instala ipywidgets: pip install ipywidgets")
    print("Alternativamente, los gráficos de Plotly ya son interactivos por defecto.")

# Nota: Para dashboards complejos con múltiples callbacks y layouts avanzados,
# es mejor crear una aplicación Dash standalone fuera del notebook.
````
