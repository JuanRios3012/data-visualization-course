
# Módulo 1: Cimientos y Anatomía de un Gráfico (Matplotlib)

En este primer módulo aprenderemos sobre la **gramática de los gráficos** en Python usando la biblioteca Matplotlib. Comprender la anatomía de un gráfico te permitirá crear visualizaciones claras y precisas desde el principio.  

Un gráfico en Matplotlib está compuesto por varios elementos principales:

* **Figura** (`Figure`): es el contenedor general donde se dibujan todos los elementos de la visualización.  
* **Ejes** (`Axes`): son las áreas dentro de la figura donde se representan los datos; cada eje contiene su sistema de coordenadas (ejes *x* y *y*).  
* **Artistas**: son los objetos visibles (líneas, puntos, parches, textos, leyendas) que Matplotlib dibuja dentro de los ejes.  

A lo largo de este módulo explorarás cómo crear figuras y ejes, dibujar diferentes tipos de gráficos básicos (líneas, barras, dispersión e histogramas) y personalizarlos con títulos, etiquetas, leyendas, anotaciones, escalas y estilos.

## Gramática de los gráficos: Figura y Ejes

Matplotlib grafica los datos sobre *figuras* que pueden contener uno o más *ejes*.  
Cada eje define un sistema de coordenadas y dispone de métodos para añadir datos (como `plot()`, `scatter()`, `bar()`), ajustar límites y escalas, y añadir etiquetas o leyendas.  

- Una **Figura** es un contenedor de alto nivel que agrupa uno o varios ejes y otros artistas especiales (títulos, barras de color, etc.).  
- Un **Eje** es el área dentro de la figura donde se dibujan los datos; cada eje suele tener dos objetos `Axis` (eje *x* y eje *y*) que crean las escalas y marcan las etiquetas.  
- El método más sencillo de crear una figura con un eje es `plt.subplots()`, que devuelve la figura y uno o varios ejes, sobre los que podemos dibujar datos con métodos como `ax.plot()` y mostrar la figura con `plt.show()`.  
- En la interfaz orientada a objetos, instanciamos explícitamente `Figure` y `Axes` y llamamos a sus métodos; la interfaz *pyplot* crea estas instancias de manera implícita.

Este modelo de objetos nos permite un control preciso sobre cada elemento de la visualización.



```python
# Importamos Matplotlib y Numpy
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Creamos una figura y un eje explícitamente (estilo orientado a objetos)
fig, ax = plt.subplots()

# Datos simples para trazar
t = np.linspace(0, 10, 100)
y = np.sin(t)

# Dibujamos una línea en el eje
ax.plot(t, y)

# Añadimos un título y etiquetas de ejes
ax.set_title('Ejemplo de curva seno')
ax.set_xlabel('t')
ax.set_ylabel('sin(t)')

# Mostramos la figura (en notebooks no es necesario llamar a plt.show())
plt.show()
```


### Interfaces de Matplotlib

Matplotlib ofrece dos estilos de uso:

- **Interfaz orientada a objetos (OO)**: tú creas explícitamente una figura y uno o más ejes (`fig, ax = plt.subplots()`), y luego llamas a los métodos de `ax` para dibujar y personalizar el gráfico.  
- **Interfaz pyplot (estilo MATLAB)**: utilizas funciones de `matplotlib.pyplot` como `plt.plot()`, `plt.title()` y `plt.xlabel()` que crean y gestionan la figura y los ejes de forma implícita.  

Por ejemplo, el siguiente código utiliza la interfaz *pyplot* para trazar la misma curva seno de la celda anterior:


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('Interfaz pyplot: seno y coseno')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()  # Mostramos una leyenda

plt.show()
```


## Gráficos de líneas

Los gráficos de líneas son una forma sencilla de visualizar la relación entre una variable independiente y una dependiente. Puedes personalizar el color, estilo de línea y marcador de los puntos utilizando argumentos de palabra clave en `plt.plot()` o `ax.plot()`:

- **Color (`color`)**: puedes especificar colores por nombre (`'blue'`), códigos abreviados (`'g'` para verde), valores de gris (`'0.75'`), códigos hexadecimales (`'#FFDD44'`) o tuplas RGB.  
- **Estilo de línea (`linestyle`)**: valores como `'solid'`, `'dashed'`, `'dashdot'` o `'dotted'` controlan el patrón de la línea. También puedes usar códigos abreviados (`'-'`, `'--'`, `'-.`', `':'`).  
- **Marcadores (`marker`)**: caracteres como `'o'`, `'x'`, `'s'`, etc., definen la forma de los puntos cuando quieres resaltar cada observación.  

A continuación se muestra un ejemplo con varias combinaciones.


```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)

#linestyles: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
#colores: https://matplotlib.org/stable/gallery/color/named_colors.html, https://htmlcolorcodes.com/es/

plt.plot(x, np.sin(x - 0), color='blue', linestyle='-', label='Azul sólido') 
plt.plot(x, np.sin(x - 1), color='g', linestyle='--', label='Verde discontinuo')
plt.plot(x, np.sin(x - 2), color='0.75', linestyle='-.', label='Gris (punto y raya)')
plt.plot(x, np.sin(x - 3), color='#FF5733', linestyle=':', label='Rojo punteado') 

# Títulos y leyenda
plt.title('Personalización de líneas')
plt.xlabel('x')
plt.ylabel('sin(x + fase)')
plt.legend()

plt.show()
```


## Gráficos de dispersión

Los gráficos de dispersión representan pares de valores como puntos individuales. Se pueden crear con `plt.plot(x, y, 'o')` o con `plt.scatter()`.  
Usar `plt.scatter()` permite asignar **colores** y **tamaños** diferentes a cada punto, lo que resulta útil para visualizar más de dos dimensiones de información.  

También puedes especificar distintos marcadores pasando caracteres como `'o'`, `'.'`, `','`, `'x'`, `'v'`, etc. a la función `plot()`.  
A continuación se muestran ejemplos de ambos enfoques.


```python

import matplotlib.pyplot as plt
import numpy as np

# Datos para ejemplos sencillos
x = np.linspace(0, 10, 30)
y = np.sin(x)

# Scatter simple con markers diferentes
plt.figure(figsize=(6, 4))
plt.plot(x, y, 'o', color='black', label="'o' (círculos)")

# Varios marcadores en un mismo gráfico
rng = np.random.RandomState(0)
plt.figure(figsize=(6, 4))
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label=marker)
plt.legend(numpoints=1, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Ejemplos de marcadores')
plt.tight_layout()

# Scatter con tamaño y color dependientes de los datos
rng = np.random.RandomState(1)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.figure(figsize=(6, 4))
#color palletes https://matplotlib.org/stable/users/explain/colors/colormaps.html
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar(label='Valor de color')
plt.title('Scatter con colores y tamaños variables')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
```


## Gráficos de barras

Los gráficos de barras son adecuados para comparar cantidades categóricas. La función `plt.bar()` toma como argumentos una serie de categorías (posiciones en el eje *x*) y sus valores. Puedes personalizar el color de cada barra y añadir etiquetas en la parte superior.  

A continuación, un ejemplo con datos ficticios de ventas por producto.


```python

import matplotlib.pyplot as plt

# Datos categóricos
y_categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
y_values = [25, 40, 30, 15]
colors = ['steelblue', 'orange', 'green', 'red']

# Creamos la figura y el eje
fig, ax = plt.subplots()
ax.bar(y_categories, y_values, color=colors)

# Añadimos etiquetas y valores sobre cada barra
for idx, value in enumerate(y_values):
    ax.text(idx, value + 0.5, str(value), ha='center')

ax.set_title('Ventas por producto')
ax.set_ylabel('Ventas (en miles)')

# plt.tight_layout()
plt.show()
```


## Histogramas

Un histograma representa la distribución de una variable numérica dividiendo su rango en *bins* (intervalos) y contando cuántos valores caen en cada bin.  
La función `plt.hist()` permite ajustar el número de bins, la transparencia (`alpha`) y el tipo de trazo (`histtype`) para comparar varias distribuciones.  

A continuación se muestra un ejemplo sencillo y luego un histograma múltiple.


```python

import matplotlib.pyplot as plt
import numpy as np

# Un conjunto de datos aleatorios
data = np.random.randn(1000)

plt.figure(figsize=(6, 3))
plt.hist(data, bins=20, alpha=0.5, color='steelblue', edgecolor='black')
plt.title('Histograma simple')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.tight_layout()

# Histograma múltiple para comparar tres distribuciones
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.figure(figsize=(6, 3))
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=30)
plt.hist(x1, **kwargs, color='red', label='Distribución 1')
plt.hist(x2, **kwargs, color='green', label='Distribución 2')
plt.hist(x3, **kwargs, color='blue', label='Distribución 3')
plt.title('Histogramas comparativos')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.tight_layout()
```


## Etiquetas, títulos y leyendas

Para que los gráficos sean comprensibles, hay que añadir títulos y etiquetas a los ejes, y una leyenda cuando se muestran varias series.  
Los objetos `Axes` proporcionan métodos como `set_title()`, `set_xlabel()`, `set_ylabel()` y `legend()` para este fin.  
En la interfaz *pyplot* los equivalentes son `plt.title()`, `plt.xlabel()`, `plt.ylabel()` y `plt.legend()`.  

El ejemplo siguiente muestra cómo etiquetar dos series y colocar la leyenda fuera del área de dibujo.


```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)
y1 = np.cumsum(np.random.randn(200))
y2 = np.cumsum(np.random.randn(200))

fig, ax = plt.subplots()
ax.plot(x, y1, label='Serie 1', color='purple')
ax.plot(x, y2, label='Serie 2', color='teal')

ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Valor acumulado')
ax.set_title('Ejemplo con leyenda')

# Colocamos la leyenda en la esquina superior izquierda, fuera del gráfico
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```


## Anotaciones

Las anotaciones sirven para destacar puntos o áreas específicas del gráfico mediante texto y flechas.  
La función `annotate()` permite posicionar el texto en coordenadas del gráfico y dibujar opcionalmente una flecha que apunte al punto de interés.  

El siguiente ejemplo señala el máximo local de una función seno.


```python

import matplotlib.pyplot as plt
import numpy as np

# Datos de la función
x = np.linspace(0, 5, 100)
y = np.cos(2 * np.pi * x)

fig, ax = plt.subplots()
ax.plot(x, y)

# Anotación: texto y flecha señalando un máximo local
ax.annotate('Máximo local', xy=(1, 1), xytext=(1.5, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_title('Anotación en un gráfico')
ax.set_ylim(-2, 2)

plt.show()
```


## Escalas y límites de los ejes

Los ejes proporcionan métodos para ajustar los límites y la escala de las coordenadas. Por ejemplo, `set_xlim()` y `set_ylim()` modifican el rango visible de los datos, y `set_yscale('log')` cambia la escala a logarítmica.

También puedes manipular los *ticks* (marcas y etiquetas en los ejes) con `set_xticks()`, `set_yticks()` o el método `tick_params()` para cambiar color, posición y otros atributos.

A continuación se muestra un gráfico con escala logarítmica y límites personalizados.


```python
import matplotlib.pyplot as plt
import numpy as np

# Simulamos un camino aleatorio exponencial
np.random.seed(0)
steps = np.random.randn(200)
x = np.arange(len(steps))
y = np.exp(np.cumsum(steps))  # valores crecen exponencialmente

fig, ax = plt.subplots()
ax.plot(x, y)

# Cambiamos la escala del eje y a logarítmica y ajustamos límites
ax.set_yscale('log')
ax.set_xlim(20, 180)
ax.set_xlabel('Índice')
ax.set_ylabel('Valor acumulado (escala log)')
ax.set_title('Camino aleatorio con escala logarítmica')

# Personalizamos los ticks: ponemos ticks menores y coloreamos
ax.tick_params(top=True, labeltop=True, color='red', labelcolor='green')

plt.tight_layout()
plt.show()
```


## Estilos y configuración global (rcParams)

Matplotlib permite personalizar la apariencia global de los gráficos mediante tres mecanismos:

1. **rcParams en tiempo de ejecución**: un diccionario que guarda la configuración actual. Puedes modificar parámetros como el ancho de línea (`lines.linewidth`), estilo (`lines.linestyle`) o la paleta de colores (`axes.prop_cycle`) directamente en este diccionario.
  
2. **Hojas de estilo** (`style.use()`): ficheros que contienen una colección de parámetros; Matplotlib ofrece estilos predefinidos como `'ggplot'` o `'seaborn'`.  
3. **Archivo `matplotlibrc`**: archivo de configuración permanente que se carga en cada sesión.

Los parámetros definidos en `rcParams` tienen prioridad sobre las hojas de estilo, y estas a su vez sobre el fichero `matplotlibrc`.


```python

import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

# Datos para el ejemplo
t = np.linspace(0, 2*np.pi, 400)

# Gráfico con configuración predeterminada
g1 = np.sin(t)
g2 = np.cos(t)

fig, ax = plt.subplots()
ax.plot(t, g1, label='sin(t)')
ax.plot(t, g2, label='cos(t)')
ax.set_title('Estilo predeterminado')
ax.legend()
plt.show()

# Aplicamos un estilo predefinido ('ggplot')
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot(t, g1, label='sin(t)')
ax.plot(t, g2, label='cos(t)')
ax.set_title("Estilo 'ggplot'")
ax.legend()
plt.show()

# Personalizamos rcParams temporalmente
with mpl.rc_context({'lines.linewidth': 4, 'lines.linestyle': '-.'}):
    fig, ax = plt.subplots()
    ax.plot(t, g1, label='sin(t) modificado')
    ax.plot(t, g2, label='cos(t) modificado')
    ax.set_title('Personalización con rc_context')
    ax.legend()
plt.show()
```


## Subplots y organización de múltiples gráficos

Para comparar varios gráficos dentro de la misma figura, Matplotlib permite crear rejillas de subgráficos mediante `plt.subplots()` o `plt.subplot_mosaic()`.  
`plt.subplots(nrows, ncols)` devuelve una figura y un arreglo de ejes para una cuadrícula de tamaño (`nrows`, `ncols`), mientras que `plt.subplot_mosaic()` permite asignar nombres a cada subgráfico y crear disposiciones más flexibles.  

En el siguiente ejemplo se utilizan cuatro subgráficos para mostrar distintas funciones.


```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 400)

# Creamos una figura con una cuadrícula 2x2 de ejes
fig, axs = plt.subplots(2, 2, figsize=(8, 6), layout='constrained')

# Dibujamos diferentes funciones en cada subgráfico
axs[0, 0].plot(x, np.sin(x), color='navy')
axs[0, 0].set_title('sin(x)')

axs[0, 1].plot(x, np.cos(x), color='crimson')
axs[0, 1].set_title('cos(x)')

axs[1, 0].plot(x, np.tan(x), color='green')
axs[1, 0].set_ylim(-10, 10)  # limitamos para evitar valores grandes
axs[1, 0].set_title('tan(x)')

axs[1, 1].plot(x, np.sin(x) * np.cos(x), color='orange')
axs[1, 1].set_title('sin(x)·cos(x)')

# Añadimos un título general a la figura
fig.suptitle('Ejemplos de subgráficos', fontsize=16)

plt.show()
```
